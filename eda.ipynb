```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
```

```python
#Fit Decision Tree Regressor to the dataset.
from sklearn.model_selection import train_test_split

pd.set_option('display.max_columns', 20)

feature_name = 'ShannonE.Median'
#target_name = 'count'
target_name = 'lq_all'

X = top[feature_name].values
Y = top[target_name].values

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=30)
print(X_train)

X
```

```python
#Decision Tree

d_reg = DecisionTreeRegressor(max_depth=3)#min_samples_leaf=10)  
d_reg.fit(X_train.reshape(-1,1), Y_train.reshape(-1,1))

```

```python
X_grid = np.arange(0, 0.7, 0.01)
X_grid = X_grid.reshape((len(X_grid), 1)) 
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
top[['Sobs.Median','Chao1.Median', 'Goods.Median','ShannonH.Median','ShannonE.Median', 'count']]
```

```python
pca = PCA(n_components=6)
comps = pca.fit_transform(top[['Sobs.Median', 'Chao1.Median','Goods.Median',	'ShannonH.Median','ShannonE.Median', 'count']])
print(pca.explained_variance_ratio_)
```

```python
fig, ax = plt.subplots(figsize=(10,9))
plt.scatter(comps[:,0], comps[:,1], alpha=0.5, c='green')
plt.title("PCA Random Bacterial Stuff PCA")
plt.style.use('ggplot')
```
